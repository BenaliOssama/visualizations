```python
import matplotlib.pyplot as plt
```

```python
x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]
```

```python
plt.plot(x, y, '-.r', linewidth=3)
plt.plot(x, y, 'ob', markersize=12)

plt.title('Display marker')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(1, 8)
plt.ylim(1, 8)

plt.show()


```
