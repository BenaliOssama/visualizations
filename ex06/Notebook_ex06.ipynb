```python
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
returns = np.random.randn(50)
price = 100 + np.cumsum(returns)

dates = pd.date_range(start='2020-09-01', periods=50, freq='B')
df = pd.DataFrame(zip(dates, price),
                  columns=['Date','Company_A'])
```

```python
# Plotly Express
```

```python
fig = px.line(
    df,
    x='Date',
    y='Company_A',
    title='Company_1 stock price',
    labels={'Date': 'Date', 'Company_A': 'Company_A'}
)

fig.show()
```

```python
# Graph Object // more control // low level
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Date'], y=df['Company_A'])) #"here are some x and y values, decide how to display them." 
fig.update_layout(
    title='Company_A stock price',
    xaxis_title='Date',
    yaxis_title='Company_A'
)

fig.show()
```
